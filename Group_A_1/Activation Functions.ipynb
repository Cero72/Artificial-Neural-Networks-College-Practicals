```python
import numpy as np
import matplotlib.pyplot as plt
```

# Binary Step Function

```python
def binary_step(x):
    return np.heaviside(x, 1)
```

```python
x = np.linspace(-10, 10)
plt.plot(x, binary_step(x))
plt.title("binary function")
plt.show()
```

# Sigmoid Function

```python
def sigmoid(x):
    return (1/(1 + np.exp(-x)))
```

```python
x = np.linspace(-10, 10)
plt.plot(x, sigmoid(x))
plt.title("Sigmoid function")
plt.show()
```

# Linear Function

```python
def linear(x):
    return x
```

```python
x = np.linspace(-10, 10)
plt.plot(x, linear(x))
plt.title("Linear function")
plt.show()
```

# Tanh Function

```python
def tanh(x):
    return np.tanh(x)
```

```python
x = np.linspace(-10, 10)
plt.plot(x, tanh(x))
plt.title("Tanh function")
plt.show()
```

# ReLU Function

```python
def Relu(x):
    x1 = []
    for i in x:
        if i < 0:
            x1.append(0)
        else:
            x1.append(i)
    return x1
            
```

```python
x = np.linspace(-10, 10)
plt.plot(x, Relu(x))
plt.title("ReLU function")
plt.show()
```
